```python
import numpy as np

def step_activation(x):
    return np.where(x >= 0, 1, 0)

x = float(input("Enter a value: "))
print("Output:", step_activation(x))
```

```python
def perceptron(x1, x2, w1, w2, bias):
    # Calculate weighted sum
    net = (x1 * w1) + (x2 * w2) + bias

    # Step activation function
    if net >= 0:
        return 1
    else:
        return 0

# Inputs
x1 = int(input("Enter x1 (0 or 1): "))
x2 = int(input("Enter x2 (0 or 1): "))

# Weights and Bias
w1 = float(input("Enter weight w1: "))
w2 = float(input("Enter weight w2: "))
bias = float(input("Enter bias: "))

# Output
output = perceptron(x1, x2, w1, w2, bias)

print("Binary Output:", output)
```

```python
# Step Activation Function
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

# Generic Perceptron Function
def perceptron(x1, x2, w1, w2, bias):
    net = x1 * w1 + x2 * w2 + bias
    return step_function(net)

# Truth Table Function
def print_truth_table(gate_name, w1, w2, bias):
    print("\n", gate_name, "Gate")
    print("X1 X2 Output")
    print("------------")

    for x1 in [0, 1]:
        for x2 in [0, 1]:
            output = perceptron(x1, x2, w1, w2, bias)
            print(x1, " ", x2, "   ", output)

# AND Gate
print_truth_table("AND", 1, 1, -1.5)

# OR Gate
print_truth_table("OR", 1, 1, -0.5)

# NAND Gate
print_truth_table("NAND", -1, -1, 1.5)

# NOR Gate
print_truth_table("NOR", -1, -1, 0.5)
```
